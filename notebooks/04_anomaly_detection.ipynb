```python
import pandas as pd
from sklearn.ensemble import IsolationForest
```

```python
df = pd.read_csv("../data/processed/cleaned_employee_activity.csv")
```

```python
features = df[['failed_login_attempts','files_accessed','sensitive_files_accessed','download_count','session_duration_hours','risk_score']]
```

```python
model = IsolationForest(contamination=0.2, random_state=42)
df['anomaly'] = model.fit_predict(features)
```

```python
df['anomaly'] = df['anomaly'].apply(lambda x: "Anomaly" if x == -1 else "Normal")
```

```python
df[['employee_name','risk_score','anomaly']]
```

```python
df.to_csv("../data/processed/cleaned_employee_activity.csv", index=False)
```
