```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/cleaned_employee_activity.csv")
df.head()
```

```python
plt.figure()
df['department'].value_counts().plot(kind='bar')
plt.title("Logins by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
plt.hist(df['failed_login_attempts'])
plt.title("Failed Login Attempts Distribution")
plt.xlabel("Attempts")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
plt.boxplot(df['session_duration_hours'])
plt.title("Session Duration Distribution")
plt.ylabel("Hours")
plt.show()
```

```python
df['after_hours'].value_counts().plot(kind='bar')
plt.title("After Hours Activity")
plt.show()
```

```python
corr = df[['failed_login_attempts','files_accessed','sensitive_files_accessed','download_count','session_duration_hours']].corr()

plt.figure()
sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()
```
